```python
import os
import pandas as pd

# 指定文件夹路径
folder_path = 'F:\\Python\\PPG'  # 请将路径替换为你的文件夹路径

# 列出文件夹内的所有文件
files = os.listdir(folder_path)

# 循环处理每个文件
for file in files:
    file_path = os.path.join(folder_path, file)
    
    # 检查文件扩展名是.csv或.xlsx
    if file.endswith('.csv') or file.endswith('.xlsx'):
        print(f"清空文件: {file}")
        
        # 读取文件时指定编码为 'cp1252' (ANSI 编码)
        if file.endswith('.csv'):
            df_c = pd.read_csv(file_path, encoding='cp1252')
        
        # 清空数据
        df_c = pd.DataFrame(columns=df_c.columns)
        
        # 保存修改
        if file.endswith('.csv'):
            df_c.to_csv(file_path, index=False, encoding='cp1252')
        else:
            df_c.to_excel(file_path, index=False)
        
        print(f"{file} 的数据已清空")
```

```python
import csv

# 定義要處理的 CSV 檔案名稱稱
input_file = 'F:\\Python\\PPG\\output\\feature_points.csv'

# 讀取 CSV 檔案並過濾資料行
with open(input_file, mode='r', newline='') as infile:
    reader = csv.reader(infile)
    rows = [row for row in reader if len(row) >= 13]  # 保留列數大於或等於 14 的行

# 將過濾後的資料寫入到新的 CSV 檔案中
with open(input_file, mode='w', newline='') as outfile:
    writer = csv.writer(outfile)
    writer.writerows(rows)

print(f"已經完成過濾，結果已保存到 {input_file}")
```

```python
import os
import shutil
import csv

# 定義 CSV 文件名稱和資料夾路徑
csv_file = 'F:\\Python\\PPG\\output\\feature_points.csv'  # 替換成你過濾後的 CSV 文件名稱
source_folder = 'F:\\Python\\PPG\\output\\bad signal'  # 資料夾內含有檔案的來源資料夾
destination_folder = 'F:\\Python\\PPG\\output\\filtered_files'  # 儲存匹配檔案的新資料夾

# 如果目標資料夾不存在，則創建它
if not os.path.exists(destination_folder):
    os.makedirs(destination_folder)

# 從 CSV 文件中提取所有名字
with open(csv_file, mode='r', newline='') as file:
    reader = csv.reader(file)
    file_names = [row[0] for row in reader if row]  # 取得每行的第一個值作為檔案名稱

# 在資料夾中找到與 CSV 名稱匹配的檔案並保存
for file_name in file_names:
    # 確保文件存在於來源資料夾中
    source_file_path = os.path.join(source_folder, file_name)
    if os.path.exists(source_file_path):
        destination_file_path = os.path.join(destination_folder, file_name)
        shutil.copy(source_file_path, destination_file_path)
        print(f"已保存 {file_name} 到 {destination_folder}")
    else:
        print(f"未找到 {file_name} 在 {source_folder}")

print("檔案保存完成。")
```

```python
import pandas as pd

# 定義 CSV 文件名稱
csv_file = 'F:\\Python\\PPG\\output\\feature_points.csv'  # 替換成你的 CSV 文件名稱

# 讀取 CSV 文件
df = pd.read_csv(csv_file)

# 將所有數字列中的值加上 5
df = df.applymap(lambda x: x + 5 if isinstance(x, (int, float)) else x)

# 將結果保存回 CSV 文件
df.to_csv(csv_file, index=False)

print(f"已經完成將 CSV 中所有數字加上 5，結果已保存到 {csv_file}")
```

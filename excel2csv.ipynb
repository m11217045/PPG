```python
import pandas as pd
origin_train = r"F:\\Excel\\train.xlsx"
exc_path_train = r"F:\\Excel\\train_Dataset.xlsx"
csv_path_train = r"F:\\Excel\\train_csv.csv"

origin_test = r"F:\\Excel\\test.xlsx"
exc_path_test = r"F:\\Excel\\test_Dataset.xlsx"
csv_path_test = r"F:\\Excel\\test_csv.csv"

pathList = [origin_train,origin_test]
pathEXC = [exc_path_train,exc_path_test]
pathCSV = [csv_path_train,csv_path_test]
```

```python
for i in range(len(pathList)):
    print(pathEXC[i])
    s1 = pd.read_excel(pathList[i], sheet_name='Sheet1',usecols="A,B,C,D,Z,AB,AD,AF")
    s2 = pd.read_excel(pathList[i], sheet_name='Sheet2',usecols="Z,AB,AD,AF")
    s3 = pd.read_excel(pathList[i], sheet_name='Sheet3',usecols="Z,AB,AD,AF")
    s4 = pd.read_excel(pathList[i], sheet_name='Sheet4',usecols="B,C,E,F,H,I,K,L,N,P,Q,R,T,V,W,X,Z,AB,AD,AF,AG,AH,AJ,AK,AL,AN,AP,AQ,AR,AT,AV,AX,AZ,BA,BB")

    df_c = pd.concat([s1, s2, s3, s4], axis=1, ignore_index=True)
    column_name = ("NAME","DATE","HAND","STATE","HH_AVG","HA_AVG","AH_AVG","AA_AVG","HH_STD","HA_STD","AH_STD","AA_STD","HH_VAR","HA_VAR","AH_VAR","AA_VAR","HH_positive","HH_negative","HA_positive","HA_negative","AH_positive","AH_negative","AA_negative","Aanegative","HR","HH_上升AVG","HH_上升STD","HH_上升VAR","HH上升斜率","HH下降AVG","HH下降STD","HH下降VAR","HH下降斜率","HH最大寬度","HHPPGA","HHSSI_AVG","HHSSI_STD","HHSSI_VAR","AA_上升AVG","AA_上升STD","AA_上升VAR","AA上升斜率","AA下降AVG","AA下降STD","AA下降VAR","AA下降斜率","AA最大寬度","AAPPGA","AASSI_AVG","AASSI_STD","AASSI_VAR")
    df_c.columns = column_name
    print(df_c)
    # 將數據寫入CSV檔案
    csv_file = pathCSV[i]
    df_c.to_csv(csv_file, index=False, encoding='ANSI')
    
    excel_file = pathEXC[i]
    df_c.to_excel(excel_file, index=False, engine='openpyxl')

    print('Excel數據已成功寫入檔案。')
```

```python
import os
import pandas as pd

# 指定文件夹路径
folder_path = 'F:\\Python\\PPG'  # 请将路径替换为你的文件夹路径

# 列出文件夹内的所有文件
files = os.listdir(folder_path)

# 循环处理每个文件
for file in files:
    file_path = os.path.join(folder_path, file)
    
    # 检查文件扩展名是.csv或.xlsx
    if file.endswith('.csv') or file.endswith('.xlsx'):
        print(f"清空文件: {file}")
        
        # 读取文件时指定编码为 'cp1252' (ANSI 编码)
        if file.endswith('.csv'):
            df_c = pd.read_csv(file_path, encoding='cp1252')
        
        # 清空数据
        df_c = pd.DataFrame(columns=df_c.columns)
        
        # 保存修改
        if file.endswith('.csv'):
            df_c.to_csv(file_path, index=False, encoding='cp1252')
        else:
            df_c.to_excel(file_path, index=False)
        
        print(f"{file} 的数据已清空")
```
